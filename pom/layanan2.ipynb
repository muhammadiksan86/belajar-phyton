```python
# Membaca file CSV dengan pemisah titik koma
data = pd.read_csv("data2.csv", sep=';')

# Menambah kolom "Lama Pelayanan" dengan hasil pengurangan
data['Lama Pelayanan'] = data['Selesai Pelayanan'] - data['Mulai Pelayanan']
# Menambah kolom "Pelanggan Menunggu" dengan hasil pengurangan
data['Pelanggan Menunggu'] = data['Mulai Pelayanan'] - data['Waktu Kedatangan']
# Menambah kolom "Petugas idle" dengan hasil pengurangan
data['Petugas idle'] = data['Mulai Pelayanan'] - data['Selesai Pelayanan'].shift()

# Baris Total untuk "Lama Pelayanan"
total1 = data['Lama Pelayanan'].sum()

# Baris Total untuk "Pelanggan Menunggu"
total2 = data['Pelanggan Menunggu'].sum()

# Baris Total untuk "Petugas idle"
total3 = data['Petugas idle'].sum()

# Membuat satu baris total dengan semua total di atas
total_row = pd.DataFrame({"Pelanggan": "Total", 
                         "Waktu Kedatangan": "",
                         "Mulai Pelayanan": "",
                         "Selesai Pelayanan": "",
                         "Lama Pelayanan": total1,
                         "Pelanggan Menunggu": total2,
                         "Petugas idle": total3}, index=[len(data)])

# Menggabungkan data frame asli dengan baris total
data = pd.concat([data, total_row])

# Menampilkan data frame
data
```
